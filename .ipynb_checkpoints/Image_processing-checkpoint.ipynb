# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = 'retina'
from matplotlib import rcParams
```

# Image Visualization

```python
covid = plt.imread('COVID/Covid (2).png')#Let's import a covid positive scan
noncovid = plt.imread('non-COVID/Non-Covid (5).png')#Let's import a covid positive scan

fig, ax = plt.subplots(1,2,figsize = (10,3),constrained_layout=True)
#plt.axis("off")
ax[0].imshow(covid)
ax[0].set_title('Covid Positive CT scan') 
ax[1].imshow(noncovid)
ax[1].set_title('Covid Negative CT scan') 
plt.show()
```
